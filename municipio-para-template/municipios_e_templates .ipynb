## Leitura dos dados

```python
arq = open("municipios.txt")
municipios = arq.read()
arq.close()
```

## Criação da tabela de dados

```python
municipios = municipios.split("\n")
```

```python
dic = {"adpmnet": "ADPM", "facil": "Portal Facil", "memory": "Memory", "pm-": "Siplanweb", "portatp": "Portal TP", "portalcidadao": "GRP"}
```

```python
data = []

for mun in municipios:
    aux = mun.split(";")
    added = False
    for key, value in dic.items():
        if key in aux[1]:
            data.append([aux[0].replace("\t", ""), aux[1].replace("\t", ""), value])
            added = True

    if not added:
        data.append([aux[0].replace("\t", ""), aux[1].replace("\t", ""), ""])

  
```

## Visualização usando Pandas

```python
import pandas as pd

df = pd.DataFrame(data, columns = ["Municipio", "Url", "Template"])
```

## Consultas

```python
def buscaPorTemplate(template):

    return df[df["Template"] == template]
```

```python
buscaPorTemplate("ADPM")
```

```python
def buscaPorCidade(cidade):

    return df[df["Municipio"] == cidade]
```
